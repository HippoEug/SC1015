### Dataset from "year" Summer Olympics medal table wiki

```python
# Importing required library
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 

sb.set() # Setting default seaborn style as graphics
```

#### Need to loop so that is display other years

```python
for years in range(2000, 2020, 4):
    html_link = 'https://en.wikipedia.org/wiki/' + str(years) + '_Summer_Olympics_medal_table'
    medal_over_years = pd.read_html(html_link)
    print(years)
    #print(html_link)
    #print("Data type : ", type(medal_over_years))
    #print("HTML tables : ", len(medal_over_years))
    
    medal_table = pd.DataFrame(medal_over_years[2])
    #print(medal_table.shape)
    display(medal_table.head(20))
```

#### Display Top 20 countries from each year stated

```python
for years in range(2000, 2020, 4):
    html_link = 'https://en.wikipedia.org/wiki/' + str(years) + '_Summer_Olympics_medal_table'
    medal_over_years = pd.read_html(html_link)
    print(years)
    #print(html_link)
    #print("Data type : ", type(medal_over_years))
    #print("HTML tables : ", len(medal_over_years))
    
    medal_table = pd.DataFrame(medal_over_years[2])
    #print(medal_table.shape)
    #display(medal_table.head(20))
    top_country = pd.DataFrame(medal_table.iloc[0:20, 1:2])
    # iloc: Purely integer-location based indexing for selection by position. [rows, columns]
    display(top_country.head(20))
```
