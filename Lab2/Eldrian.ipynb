# Problem 1: Data Preparation

Import Python Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Question a

Read csv and show train DataFrame (index_col gets the column Id as the row header instead of creating its own) 

```python
train_df = pd.read_csv("train.csv", index_col="Id")
train_df
```

## Question b

Show columns dtypes

```python
train_df.dtypes
```

## Question c

Get all columns of dtype "int64"

```python
int_train_df = train_df.select_dtypes(include="int64")
display(int_train_df)
```

## Question d and e

Remove categorical columns

```python
clean_df = int_train_df.drop(columns=["MSSubClass", "OverallQual", "OverallCond", "MoSold"])
clean_df
```

# Problem 2 : Statistical Summary

## Question a

Summary Statistics of SalePrice

```python
saleprice = pd.DataFrame(clean_df["SalePrice"])
saleprice.describe()
```

## Question b

Using Seaborn

```python
f = plt.figure(figsize=(24,6))
sns.boxplot(data=saleprice, orient="h")
```

Using Pandas Function(Technically it's matplot)

```python
clean_df.boxplot(column="SalePrice", vert=False, figsize=(24,6))
```

Using both Seaborn and Matplot

```python
f = plt.figure(figsize=(30,6))
sns.histplot(data=saleprice)
clean_df.hist(column="SalePrice", figsize=(24,6), bins=50)
```

Using both Seaborn

```python
f = plt.figure(figsize=(24,6))
sns.kdeplot(data=saleprice)
```

Using Pandas function

```python
saleprice.plot.kde(figsize=(26,6))
```

## Question c

Summary statistics of LotArea

```python
clean_df["LotArea"].describe()
```

## Question d

```python
clean_df["LotArea"].plot.box(vert=False, figsize=(26,6), grid=True)
```

```python
lotarea = pd.DataFrame(clean_df["LotArea"])
f = plt.figure(figsize=(24,6))
sns.histplot(data=lotarea)
```

```python
clean_df["LotArea"].plot.kde(figsize=(26,6))
```

## Question e

```python
sns.jointplot(x="LotArea", y="SalePrice", data=clean_df[["LotArea", "SalePrice"]], height=20)
```
