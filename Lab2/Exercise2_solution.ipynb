# Problem 2

### a: Find Summary Statistics (Mean, Median, Quartiles etc) of SalePrice

```python
saleprice = pd.DataFrame(clean_df['SalePrice'])
saleprice.describe()
```

### b: Summary Statistics & Distribution of SalePrice using standard Box-Plot, Histogram, KDE

```python
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = saleprice, orient = "h")
```

```python
clean_df.boxplot(column="SalePrice", vert=False, figsize=(24,4))
```

```python
f = plt.figure(figsize=(16, 8))
sb.histplot(data = saleprice)
```

```python
clean_df.hist(column="SalePrice", figsize=(16,8), bins=50)
```

```python
f = plt.figure(figsize=(16, 8))
sb.kdeplot(data = saleprice)
```

```python
saleprice.plot.kde(figsize=(16,8))
```

### c: Find Summary Statistics (Mean, Median, Quartiles etc) of LotArea

```python
lotarea = pd.DataFrame(clean_df['LotArea'])
lotarea.describe()
```

### d: Summary Statistics & Distribution of LotArea using standard Box-Plot, Histogram, KDE

```python
f = plt.figure(figsize=(26, 6))
sb.boxplot(data = lotarea, orient = "h")
```

```python
f = plt.figure(figsize=(24, 6))
sb.histplot(data = lotarea)
```

```python
f = plt.figure(figsize=(26, 6))
sb.kdeplot(data = saleprice)
```

### e: Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot and find the Correlation

```python
sb.jointplot(data = clean_df[["LotArea", "SalePrice"]], x = "LotArea", y = "SalePrice", height = 20)
```
